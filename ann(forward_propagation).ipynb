XOR problem using multi layer neural network(forward propogation only)


```python
import numpy as np
```

dEFINE XOR DATASET

```python
x=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
x
```

```python
# define step activation function
def step(z):
  return 1 if z>=0 else 0
```

```python
w1,w2,b1=[1,1,-0.5]
w3,w4,b2=[1,1,-1.5]
w5,w6,b3=[1,-2,-0.5]
print("succesfull")
```

```python
def forward_propagation(x1,x2):
  h1=w1*x1+w2*x2+b1
  a_h1=step(h1)
  h2=w3*x1+w4*x2+b2
  a_h2=step(h2)
  y=w5*a_h1+w6*a_h2+b3
  output=step(y)
  return output
```

```python
print("xor results using ann forward propagation")
for X in x:
  prediction=forward_propagation(X[0],X[1])
  print(f"input: {X} output: {prediction}")
```

Perceptron (Binary Clasification)
example: study hours -> pass(1)/fail(0)

step1: preparing training data

```python
import numpy as np
```

```python
x=np.array([1,2,3,4])
y=np.array([0,0,1,1])
x,y
```

step 1: initialize parameters


```python
w=0.0
b=0.0
learning_rate=1
epochs=5
print("initial weight:",w)
print("initialise bias:",b)
```

step 2: define activation function(step function)

```python
def step_function(z):
  return 1 if z>=0 else 0
```

Training loop

```python
for epoch in range(epochs):
  print(f"\n Epoch {epoch+1}")
  for i in range(len(x)):
    # compute weighted sum
    z=w*x[i]+b
    #apply activation function
    y_pred=step_function(z)
    #compute error
    error=y[i]-y_pred
    # update weight
    w=w+learning_rate*(error)*x[i]
    #update bias
    b=b+learning_rate*(error)
    print(f"input:{x[i]},predicted:{y_pred},acctual:{y[i]},w:{w},b:{b}")

```

```python
#final model parameter
print("final weight:",w)
print("final bias:",b)
```

```python
#testing the model
# the test_value is less than < 3 then 0 , else greater than 3
test_value=3.9
z=w*test_value+b
prediction=step_function(z)
print(f"for study hours {test_value},prediction",prediction)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7, 4))

# scatter: class 0 and class 1
plt.scatter(x[y == 0], y[y == 0], marker='o', label="Class 0")
plt.scatter(x[y == 1], y[y == 1], marker='x', label="Class 1")

# Decision boundary: x = -b / w
if w != 0:
    x_boundary = -b / w
    plt.axvline(x=x_boundary, linestyle='--', label=f"Decision boundary x={x_boundary:.2f}")
else:
    print("\nDecision boundary not defined because w = 0 (vertical line can't be computed).")

plt.title("Perceptron: Data and Decision Boundary")
plt.xlabel("x (study hours)")
plt.ylabel("y (label)")
plt.yticks([0, 1])
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.legend()
plt.show()

# ---- Test the model ----
test_value = 3.9
z = w * test_value + b
prediction = step_function(z)
print(f"\nFor study hours {test_value}, prediction = {prediction}")
```

Business scenario: email spam detection for a start up

```python
import numpy as np

# Features: [SuspiciousKeywords, SenderUnknown]
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=float)

# OR labels: Spam if any feature is 1
y = np.array([0, 1, 1, 1], dtype=int)

print("X:\n", X)
print("y:", y)
```

```python
def step(z):
    return 1 if z >= 0 else 0
```

```python
w = np.zeros(X.shape[1])  # 2 weights
b = 0.0
lr = 1.0
epochs = 10

for epoch in range(epochs):
    total_errors = 0
    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        y_pred = step(z)
        error = y[i] - y_pred

        # update
        w = w + lr * error * X[i]
        b = b + lr * error

        total_errors += abs(error)

    print(f"Epoch {epoch+1}: errors={total_errors}, w={w}, b={b}")

print("\nFinal w:", w, "Final b:", b)
```

```python
for xi, yi in zip(X, y):
    z = np.dot(w, xi) + b
    pred = step(z)
    print(f"Input={xi.astype(int)}  Pred={pred}  Actual={yi}")
```

```python
new_email = np.array([1, 0], dtype=float)
z = np.dot(w, new_email) + b
pred = step(z)

print("New email features [keywords, unknown] =", new_email.astype(int))
print("Prediction (1=Spam, 0=Not spam):", pred)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))

# plot points
for label, marker in [(0, 'o'), (1, 'x')]:
    plt.scatter(X[y==label, 0], X[y==label, 1], marker=marker, s=120, label=f"Class {label}")

# decision boundary: w1*x1 + w2*x2 + b = 0
x1_vals = np.linspace(-0.5, 1.5, 200)

if w[1] != 0:
    x2_vals = -(w[0]*x1_vals + b) / w[1]
    plt.plot(x1_vals, x2_vals, linestyle='--', label="Decision boundary")
else:
    # vertical line if w2 = 0
    x_boundary = -b / w[0] if w[0] != 0 else 0
    plt.axvline(x=x_boundary, linestyle='--', label="Decision boundary")

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("Suspicious Keywords (0/1)")
plt.ylabel("Sender Unknown (0/1)")
plt.title("Perceptron for Spam Detection (OR logic)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def step(z):
    return 1 if z >= 0 else 0

def train_perceptron(X, y, lr=1.0, epochs=25, w_init=None, b_init=0.0, verbose=False):
    w = np.zeros(X.shape[1]) if w_init is None else w_init.astype(float).copy()
    b = float(b_init)

    errors_per_epoch = []

    for epoch in range(epochs):
        total_errors = 0
        for i in range(len(X)):
            z = np.dot(w, X[i]) + b
            y_pred = step(z)
            error = y[i] - y_pred

            # update
            w = w + lr * error * X[i]
            b = b + lr * error

            total_errors += abs(error)

            if verbose:
                print(f"Epoch {epoch+1}, X={X[i]}, z={z:.2f}, pred={y_pred}, actual={y[i]}, w={w}, b={b}")

        errors_per_epoch.append(total_errors)

    return w, b, errors_per_epoch

def print_boundary(w, b):
    # Decision boundary: w1*x1 + w2*x2 + b = 0
    print("\nDecision Boundary Equation:")
    print(f"{w[0]:.2f}*x1 + {w[1]:.2f}*x2 + {b:.2f} = 0")
    # Optional slope form if w2 != 0: x2 = -(w1/w2)x1 - b/w2
    if w[1] != 0:
        print(f"Equivalent: x2 = {-(w[0]/w[1]):.2f}*x1 + {-(b/w[1]):.2f}")
    else:
        print("w2 = 0, so boundary is a vertical line in x1.")

def plot_data_and_boundary(X, y, w, b, title="Perceptron Decision Boundary"):
    plt.figure(figsize=(6,5))

    # Plot points
    plt.scatter(X[y==0, 0], X[y==0, 1], marker='o', s=120, label="Not Spam (0)")
    plt.scatter(X[y==1, 0], X[y==1, 1], marker='x', s=120, label="Spam (1)")

    # Boundary: w1*x1 + w2*x2 + b = 0
    x1_vals = np.linspace(-0.5, 1.5, 200)

    if w[1] != 0:
        x2_vals = -(w[0]*x1_vals + b) / w[1]
        plt.plot(x1_vals, x2_vals, linestyle="--", label="Decision boundary")
    else:
        # vertical line if w2 == 0
        if w[0] != 0:
            x1_boundary = -b / w[0]
            plt.axvline(x=x1_boundary, linestyle="--", label="Decision boundary")
        else:
            print("Cannot plot boundary because both weights are zero.")

    plt.xlim(-0.5, 1.5)
    plt.ylim(-0.5, 1.5)
    plt.xlabel("x1 = Suspicious Keywords (0/1)")
    plt.ylabel("x2 = Sender Unknown (0/1)")
    plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.show()

def plot_errors(errors, title="Errors per Epoch"):
    plt.figure(figsize=(6,4))
    plt.plot(range(1, len(errors)+1), errors, marker='o')
    plt.xlabel("Epoch")
    plt.ylabel("Number of misclassifications")
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
# Emails as binary features:
# x1 = suspicious keywords, x2 = sender unknown
X = np.array([
    [0,0],  # E1
    [0,1],  # E2
    [1,0],  # E3
    [1,1]   # E4
], dtype=float)

# OR labels (Spam if any feature is 1)
y_or = np.array([0, 1, 1, 1], dtype=int)

# Given initial values
w0 = np.array([0.0, 0.0])
b0 = 0.0
lr = 1.0

w, b, errors = train_perceptron(X, y_or, lr=lr, epochs=15, w_init=w0, b_init=b0)

print("Final weights:", w)
print("Final bias:", b)
print("Errors per epoch:", errors)
```

```python
print_boundary(w, b)

plot_data_and_boundary(X, y_or, w, b, title="Decision Boundary (OR Spam Dataset)")

# Geometric meaning (simple print)
print("\nGeometric Explanation:")
print("- Perceptron creates a straight line (decision boundary).")
print("- Points on one side -> predicted Spam (1), other side -> Not Spam (0).")
print("- Because OR data is linearly separable, a single line can separate class 0 from class 1.")
```

```python
y_caseA = np.array([0, 1, 0, 1], dtype=int)  # depends mainly on x2

wA, bA, errorsA = train_perceptron(X, y_caseA, lr=lr, epochs=20, w_init=w0, b_init=b0)

print("CASE A (only E3->0) Final w:", wA, "Final b:", bA)
plot_errors(errorsA, "CASE A: Errors per Epoch (Usually converges)")
print_boundary(wA, bA)
plot_data_and_boundary(X, y_caseA, wA, bA, title="CASE A Boundary (E3 changed to 0)")
```

```python
y_xor = np.array([0, 1, 1, 0], dtype=int)

wB, bB, errorsB = train_perceptron(X, y_xor, lr=lr, epochs=50, w_init=w0, b_init=b0)

print("CASE B (XOR) Final w:", wB, "Final b:", bB)
plot_errors(errorsB, "CASE B (XOR): Errors per Epoch (Does NOT converge)")
print_boundary(wB, bB)
plot_data_and_boundary(X, y_xor, wB, bB, title="CASE B Boundary Attempt (XOR - Not Separable)")

print("\nCritical Thinking (XOR):")
print("- Perceptron does NOT converge because XOR is NOT linearly separable.")
print("- A single straight line cannot separate the classes.")
print("- Limitation exposed: Single-layer perceptron fails on non-linearly separable problems.")
print("- Solution: use multi-layer neural network (MLP) or add nonlinear features.")
```
